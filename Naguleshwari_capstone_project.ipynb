```python
import json
# Convert the string representation of production companies into a list of dictionaries
df['production_companies'] = df['production_companies'].apply(lambda x: json.loads(x))
```

```python
# # Initialize an empty dictionary to store production company names and their counts
production_counts = {}

# Iterate over each row in the DataFrame
for row in df.itertuples(index=False):
    # Iterate over each production company in the row
    for company in row.production_companies:
        # Extract the production company name
        company_name = company['name']
        
        # Increment the count for the production company
        if company_name in production_counts:
            production_counts[company_name] += 1
        else:
            production_counts[company_name] = 1

# Create a new DataFrame to store the production company names and their counts
production_df = pd.DataFrame({'Production Company': list(production_counts.keys()), 'Count': list(production_counts.values())})

# Sort the DataFrame by count in descending order
production_df = production_df.sort_values('Count', ascending=False)
production_df
```

# TASK10

```python
# Sort the data in descending order based on revenue and filter the top 500 movies
top_500_movies = df.nlargest(500, 'revenue')

# Calculate the measures of central tendency for budget, revenue, and runtime
budget_mean = top_500_movies['budget'].mean()
budget_median = top_500_movies['budget'].median()
budget_mode = top_500_movies['budget'].mode().values[0]

revenue_mean = top_500_movies['revenue'].mean()
revenue_median = top_500_movies['revenue'].median()
revenue_mode = top_500_movies['revenue'].mode().values[0]

runtime_mean = top_500_movies['runtime'].mean()
runtime_median = top_500_movies['runtime'].median()
runtime_mode = top_500_movies['runtime'].mode().values[0]

# Perform outlier analysis using box plots
plt.figure(figsize=(10, 6))
plt.subplot(1, 3, 1)
plt.boxplot(top_500_movies['budget'], vert=False)
plt.title('Budget Box Plot')

plt.subplot(1, 3, 2)
plt.boxplot(top_500_movies['revenue'], vert=False)
plt.title('Revenue Box Plot')

plt.subplot(1, 3, 3)
plt.boxplot(top_500_movies['runtime'], vert=False)
plt.title('Runtime Box Plot')

plt.show()

# Display the measures of central tendency
print(f"Budget Mean: {budget_mean}, Median: {budget_median}, Mode: {budget_mode}")
print(f"Revenue Mean: {revenue_mean}, Median: {revenue_median}, Mode: {revenue_mode}")
print(f"Runtime Mean: {runtime_mean}, Median: {runtime_median}, Mode: {runtime_mode}")
```
